```python
import pandas as pd
```

```python
df = pd.read_csv("data/AAPL.csv", index_col=0)
df.head()
```

```python
p_train = 0.6
p_val = 0.2
p_test = 0.2
n = len(df)
i_train = 0
i_val = int(p_train*n)
i_test = i_val + int(p_val*n)
print(n, i_train, i_val, i_test)
```

```python
df_train = df[i_train:i_val]
df_val = df[i_val:i_test]
df_test = df[i_test:]
df_val = df_val.reset_index()
df_test = df_test.reset_index()
df_val.head()
```

```python
df_train.to_csv("data/AAPL_train.csv")
df_val.to_csv("data/AAPL_val.csv")
df_test.to_csv("data/AAPL_test.csv")
```
